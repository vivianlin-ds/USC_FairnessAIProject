```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Get Data

```python
df1 = pd.read_csv("data/raw_data/healthcare-dataset-stroke-data.csv")
df1 = df1[df1['gender'] != 'Other']
df2 = pd.read_csv("data/raw_data/heart.csv")
```

## EDA

```python
# Plotting the bar chart: stroke by gender
sns.set_theme(style="darkgrid")
plt.figure(figsize=(8, 4))

label_mapping = {'1': 'Stroke', '0': 'No Stroke'}

sns.countplot(data=df1, x='gender', hue='stroke', palette='Set2')
plt.title('Gender Distribution by Stroke Status')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Stroke', labels=['No Stroke', 'Stroke'])
plt.savefig('visualizations/gender_distrib_by_stroke.png')

plt.show()
```

```python
# Plotting the bar chart: stroke by gender
sns.set_theme(style="darkgrid")
plt.figure(figsize=(8, 4))

label_mapping = {'1': 'Heart Disease', '0': 'No Heart Disease'}

ax = sns.countplot(data=df2, x='Sex', hue='HeartDisease', palette='Set2')
plt.title('Sex Distribution by Heart Disease Status')
plt.xlabel('Sex')
plt.ylabel('Count')
ax.set_xticklabels(['Male', 'Female'])
plt.legend(title='Stroke', labels=['No Heart Disease', 'Heart Disease'])
plt.savefig('visualizations/sex_distrib_by_heart_disease.png')

plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(8, 4))

# Subplot for stroke cases
plt.subplot(1, 2, 1)
sns.countplot(data=df1[df1['stroke'] == 1], x='hypertension', hue='gender', palette='Set2')
plt.title('Hypertension Distribution for Stroke Cases')
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Male', 'Female'])

# Subplot for non-stroke cases
plt.subplot(1, 2, 2)
sns.countplot(data=df1[df1['stroke'] == 0], x='hypertension', hue='gender', palette='Set2')
plt.title('Hypertension Distribution for Non-Stroke Cases')
plt.xlabel('Hypertension')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Male', 'Female'])

plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="darkgrid")
df1['gender'] = pd.Categorical(df1['gender'], categories=['Male', 'Female'], ordered=True)
plt.figure(figsize=(8, 4))

# Subplot for stroke cases
plt.subplot(1, 2, 1)
sns.countplot(data=df1[df1['stroke'] == 1], x='heart_disease', hue='gender', palette='Set2')
plt.title('Heart Disease Distribution for Stroke Cases')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Male', 'Female'])

# Subplot for non-stroke cases
plt.subplot(1, 2, 2)
sns.countplot(data=df1[df1['stroke'] == 0], x='heart_disease', hue='gender', palette='Set2')
plt.title('Heart Disease Distribution for Non-Stroke Cases')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Male', 'Female'])

plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='age', y='bmi', hue='gender', palette='Set2')
plt.title('Relationship between Age and BMI by Gender')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='age', y='avg_glucose_level', hue='gender', palette='Set2')
plt.title('Relationship between Age and Average Glucose Level by Gender')
plt.xlabel('Age')
plt.ylabel('Average Glucose Level')
plt.legend(title='Gender')
plt.grid(True)
plt.axhline(y=150, color='r', linestyle='--')
plt.show()
```
