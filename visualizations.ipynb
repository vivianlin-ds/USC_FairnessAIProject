```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
```

```python
stroke_df = pd.read_csv('data/raw_data/healthcare-dataset-stroke-data.csv')
stroke_df = stroke_df[stroke_df['gender'].str.lower() != 'other']
stroke_1 = stroke_df[stroke_df['stroke'] == 1]
stroke_0 = stroke_df[stroke_df['stroke'] == 0]
```

First plots: histograms of age across stroke dataset when stroke == 1 and stroke == 0

```python
# Set the style of seaborn
sns.set_theme(style="darkgrid")

# Plot histogram for stroke = 1
plt.figure(figsize=(10, 6))
sns.histplot(stroke_1['age'], bins=20, color='blue', kde=True)
plt.title('Histogram of Age for Stroke = 1')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('visualizations/age_hist_stroke_1.png')
plt.show()

# Plot histogram for stroke = 0
plt.figure(figsize=(10, 6))
sns.histplot(stroke_0['age'], bins=20, color='green', kde=True)
plt.title('Histogram of Age for Stroke = 0')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('visualizations/age_hist_stroke_0.png')
plt.show()
```

Second plot: bar chart of male/female counts when stroke == 1 and stroke == 0

```python
# Set the style of seaborn
sns.set_theme(style="darkgrid")

# Plot bar chart for stroke = 1
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=stroke_1, palette='Set2')
plt.title('Count of Gender for Stroke = 1')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True)
plt.savefig('visualizations/gender_bar_stroke_1.png')
plt.show()

# Plot bar chart for stroke = 0
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=stroke_0, palette='Set2')
plt.title('Count of Gender for Stroke = 0')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True)
plt.savefig('visualizations/gender_bar_stroke_0.png')
plt.show()
```
